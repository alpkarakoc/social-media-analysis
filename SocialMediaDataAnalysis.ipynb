```python
import pandas as pd

# Örnek veri oluştur
data = {
    'user_id': [1, 2, 3, 4, 5],
    'user_engagement': [23, 45, 12, 67, 34],
    'likes': [5, 12, 3, 20, 9],
    'comments': [2, 5, 1, 7, 4]
}

df = pd.DataFrame(data)

# CSV olarak kaydet
df.to_csv('social_media_data.csv', index=False)

print("CSV file created successfully.")
```

```python
data = pd.read_csv('social_media_data.csv')
```

```python
!pip install seaborn --upgrade
plt.figure(figsize=(10, 6))
sns.distplot(data['user_engagement'], kde=True)
plt.title('User Engagement Distribution')
plt.xlabel('Engagement')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='likes', y='comments', data=data)
plt.title('Likes vs Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')
plt.show()
```

```python
top_users = data.sort_values(by='likes', ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x='user_id', y='likes', data=top_users)
plt.title('Top Users by Likes')
plt.xlabel('User ID')
plt.ylabel('Number of Likes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='likes', y='comments', data=data)
plt.title('Likes vs Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')
plt.show()
```

```python
top_users = data.sort_values(by='likes', ascending=False).head(5)
plt.figure(figsize=(8, 6))
sns.barplot(x='user_id', y='likes', data=top_users)
plt.title('Top 5 Users by Likes')
plt.xlabel('User ID')
plt.ylabel('Number of Likes')
plt.show()
```

```python
data['engagement_level'] = pd.qcut(data['user_engagement'], q=3, labels=['Low', 'Medium', 'High'])
sns.countplot(x='engagement_level', data=data)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='engagement_level', y='likes', data=data)
plt.title('Likes Distribution by Engagement Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='engagement_level', y='comments', data=data)
plt.title('Comments by Engagement Level')
plt.show()
```
